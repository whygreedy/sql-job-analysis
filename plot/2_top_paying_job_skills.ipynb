```python
import pandas as pd

filepath = '/Users/cindy/Developer/SQL/sql-job-analysis/sql_result/2_top_paying_job_skills.csv'
df = pd.read_csv(filepath)
print(df)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# count occurrences for each skill
count_per_skill = df['skills'].value_counts()

# Plot the bar chart
plt.figure(figsize=(10, 6), dpi=500)
count_per_skill.plot(kind='bar', color='skyblue')
plt.title('Required Skills For Top 10 Paying Data Analyst Jobs')
plt.xlabel('Skill')
plt.ylabel('Occurrence')
plt.xticks(rotation=45, ha='right')
plt.yticks(np.arange(0, 5, step=1))
plt.tight_layout()
plt.show()
```
