```python
import matplotlib.pyplot as plt

df = pd.read_csv(filepath)

median_demand = df['demand_count'].median()
median_salary = df['avg_salary'].median()

plt.scatter(df['demand_count'], df['avg_salary'], s=100, alpha=0.8, color='skyblue')
plt.title('How BCG Matrix could tells us more about Optimal Skills for Data Analyst Jobs')
plt.xlabel('Demand Count')
plt.ylabel('Average Salary')
plt.axvline(x=median_demand, color='r', linestyle='--', linewidth=1)
plt.axhline(y=median_salary, color='r', linestyle='--', linewidth=1)

for i, txt in enumerate(df['skill']):
    plt.annotate(txt, (df['demand_count'][i], df['avg_salary'][i]), xytext=(5, -5), textcoords='offset points')

plt.grid(False)
plt.show()
```
