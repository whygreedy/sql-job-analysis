```python
import pandas as pd
import seaborn as sns

filepath = '/Users/cindy/Developer/SQL/sql-job-analysis/sql_result/7_skills_rawdata.csv'
df = pd.read_csv(filepath)

%config InlineBackend.figure_format = 'retina'
sns.boxplot(y=df['salary'])

print('Salary statistics of all Data Analyst Jobs')
print('- Median Salary: ', df['salary'].quantile(0.5))
print('- Mean Salary: ', round(df['salary'].mean(), 0))
print('- Max Salary: ', df['salary'].max())
print('- Min Salary: ', df['salary'].min())
```

```python
print('Salary statistics of top 5 demanded skills')

top_5_demanded_skills = ['sql', 'excel', 'python', 'tableau', 'r']
top_5_data = df[df['skill'].isin(top_5_demanded_skills)]

%config InlineBackend.figure_format = 'retina'

sns.boxplot(x='skill', y='salary', data=top_5_data, palette='Reds_r', order=top_5_demanded_skills)
```

```python
print('Salary statistics of top 5 paying skills')

top_5_paying_skills = ['kafka', 'pytorch', 'perl', 'tensorflow', 'cassandra']
top_5_data = df[df['skill'].isin(top_5_paying_skills)]

%config InlineBackend.figure_format = 'retina'

sns.boxplot(x='skill', y='salary', data=top_5_data, palette='Reds_r', order=top_5_paying_skills)
```
